```python
# Check for duplicated columns
def find_duplicated_columns(dataframe):
    # Create an empty set to store the names of duplicated columns
    duplicated_columns = set()

    # Get the columns of the dataframe
    columns = dataframe.columns

    # Compare each column with all other columns
    for i in range(len(columns)):
        for j in range(i + 1, len(columns)):
            # Check if the columns are equal
            if dataframe[columns[i]].equals(dataframe[columns[j]]):
                duplicated_columns.add(columns[j])

    return list(duplicated_columns)

# Find and print duplicated columns
duplicated_columns = find_duplicated_columns(df)
print("Duplicated columns:", duplicated_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

# Example categorical columns
categorical_columns = ['Gender', 'Lifestyle', 'Habit 01', 'Habit 02', 'Habit 03', 'Habit 04', 'Habit 05', 'Habit 06']

# Initialize a dictionary to hold LabelEncoder objects
label_encoder_X = {}

# Encode each categorical column
for column in categorical_columns:
    label_encoder_X[column] = LabelEncoder()
    df[column] = label_encoder_X[column].fit_transform(df[column])
```

```python
from sklearn.model_selection import train_test_split

# Define X (features) and y (target variable)
X = df.drop(columns=['Recommendation'])  # Features
y = df['Recommendation']  # Target variable
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode target variable using LabelEncoder
label_encoder_Y = LabelEncoder()
y = label_encoder_Y.fit_transform(y)
```

```python
from imblearn.over_sampling import SMOTE

# Find the number of samples in each class
class_counts = pd.Series(y).value_counts()
min_class_count = class_counts.min()

# Balance the dataset using SMOTE
# Set k_neighbors to a value less than or equal to the number of samples in the smallest class
smote = SMOTE(random_state=42, k_neighbors=min(min_class_count - 1, 5))
X_res, y_res = smote.fit_resample(X, y)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42)

# Display shapes of training and testing sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Initialize XGBClassifier
xgb_model = XGBClassifier(random_state=42)

# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [3, 5, 7],  # Maximum depth of the trees
    'learning_rate': [0.1, 0.01, 0.001]  # Learning rate
}

# Perform GridSearchCV to find the best combination of hyperparameters
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best model
best_xgb_model = grid_search.best_estimator_

# Train the best model
best_xgb_model.fit(X_train, y_train)

# Make predictions
y_pred_xgb = best_xgb_model.predict(X_test)

# Evaluate the model
print("XGBoost Classifier (Tuned):")
print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

# Initialize RandomForestClassifier with default parameters
random_forest = RandomForestClassifier(random_state=42)

# Define hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Minimum samples required to split a node
    'min_samples_leaf': [1, 2, 4]  # Minimum samples required at each leaf node
}

# Perform GridSearchCV to find the best combination of hyperparameters
grid_search = GridSearchCV(estimator=random_forest, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best model
best_random_forest = grid_search.best_estimator_

# Train the best model
best_random_forest.fit(X_train, y_train)

# Make predictions
y_pred = best_random_forest.predict(X_test)

# Evaluate the model
print("Random Forest Classifier (Tuned):")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the dataset
# Load the dataset
file_path = './Samadhi.csv'
data = pd.read_csv(file_path)  # Corrected syntax with an underscore


# Extract features (Habits) and labels (Recommendation)
habit_columns = ['Habit 01', 'Habit 02', 'Habit 03', 'Habit 04', 'Habit 05', 'Habit 06']
data['Combined Habits'] = data[habit_columns].agg(' '.join, axis=1)  # Combine all habits into one text column
X = data['Combined Habits']  # Features
y = data['Recommendation']   # Labels

# Use the top 5 most frequent recommendations
top_recommendations = y.value_counts().head(5).index
filtered_data = data[data['Recommendation'].isin(top_recommendations)]

# Update X and y for filtered data
X = filtered_data['Combined Habits']
y = filtered_data['Recommendation']

# Encode text habits to numerical features (simple dummy encoding for demonstration)
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_encoded = vectorizer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42, stratify=y)

# Train a simple Random Forest model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=top_recommendations)

# Plot the simplified confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=top_recommendations, yticklabels=top_recommendations, cbar=False)
plt.title("Simplified Confusion Matrix (Top 5 Recommendations)", fontsize=14)
plt.xlabel("Predicted Recommendation", fontsize=12)
plt.ylabel("True Recommendation", fontsize=12)
plt.tight_layout()
plt.show()
```
